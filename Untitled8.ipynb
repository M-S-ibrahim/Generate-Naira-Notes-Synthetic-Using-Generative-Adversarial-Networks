## Normalizing the data

```python

def prepare_classification_data(data_dir):
    """
    Prepares image classification data by loading, preprocessing, and labeling images from a directory.

    Args:
        data_dir (str): Path to the directory containing subdirectories of images, each subdirectory representing a class.

    Returns:
        tuple: A tuple containing two NumPy arrays:
            - images: A 4D NumPy array of preprocessed images with shape (num_images, 28, 28, 3).
            - labels: A 1D NumPy array of class labels corresponding to the images.
    """
    img_width = 28
    img_height = 28
    preprocessed_images = []
    labels = []

    for class_name in os.listdir(data_dir):
        class_dir = os.path.join(data_dir, class_name)
        if os.path.isdir(class_dir):
            for image_name in os.listdir(class_dir):
                image_path = os.path.join(class_dir, image_name)
                if os.path.isfile(image_path):
                    img = cv2.imread(image_path)
                    if img is not None:
                        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                        img = cv2.resize(img, (img_width, img_height), interpolation=cv2.INTER_AREA)
                        img = img / 255.0
                        preprocessed_images.append(img)
                        labels.append(class_name)

    images = np.array(preprocessed_images)
    labels = np.array(labels)
    return images, labels
```

# Build Neural Network

## 3.1 Import Modelling Components

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Reshape, Concatenate, UpSampling2D, Conv2D, BatchNormalization, LeakyReLU, Activation, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
```

## Generator Model

```python
def denomination_generator(latent_dim=128, num_denominations=7, embed_dim=50):
    # Latent noise vector input
    noise_input = Input(shape=(latent_dim,), name="noise_input")

    # Label input representing note denomination (e.g., 50, 100, 200, etc.)
    label_input = Input(shape=(1,), dtype='int32', name="denomination_label")

    # Embed and flatten the denomination label
    label_embedding = Embedding(input_dim=num_denominations, output_dim=embed_dim)(label_input)
    label_embedding = Flatten()(label_embedding)

    # Concatenate the noise and embedded label so the generator is conditioned on the denomination.
    combined_input = Concatenate()([noise_input, label_embedding])

    # Dense projection: project the combined vector into an 8x8x256 tensor.
    x = Dense(8 * 8 * 256)(combined_input)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Reshape((8, 8, 256))(x)

    # Upsampling Block 1: 8x8 -> 16x16
    x = UpSampling2D()(x)
    x = Conv2D(256, kernel_size=5, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)

    # Upsampling Block 2: 16x16 -> 32x32
    x = UpSampling2D()(x)
    x = Conv2D(128, kernel_size=5, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)

    # Upsampling Block 3: 32x32 -> 64x64
    x = UpSampling2D()(x)
    x = Conv2D(64, kernel_size=5, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)

    # Convolutional Refinement: Further process the feature maps for finer details.
    x = Conv2D(64, kernel_size=3, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)

    # Final Output Layer: Produce a 64x64 RGB image with tanh activation.
    x = Conv2D(3, kernel_size=3, padding='same')(x)
    output = Activation('tanh')(x)

    # Define the model with both noise and denomination label inputs.
    model = Model(inputs=[noise_input, label_input], outputs=output)
    return model
```

## Generating randon vector

```python
def generate_latent_vector(latent_dim, n_samples, n_classes=7):
    x_input = np.random.randn(latent_dim * n_samples)
    z_input = x_input.reshape(n_samples, latent_dim)
    labels = np.random.randint(0, n_classes, n_samples)
    return [z_input, labels]
```

## Predict Gen

```python
def generate_fake_samples(generator, latent_dim, n_samples):
    z_input, labels_input = generate_latent_vector(latent_dim, n_samples)
    images = generator.predict([z_input, labels_input])
    y = np.zeros((n_samples, 1))
    return [images, labels_input], y
```

## Discriminator

```python
def discriminator(img_shape=(28, 28, 3), num_denominations=8, embed_dim=50):
    # Image input branch
    image_input = Input(shape=img_shape, name="image_input")

    # Convolutional feature extraction from the image
    x = Conv2D(64, kernel_size=3, strides=2, padding='same')(image_input)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.3)(x)

    x = Conv2D(128, kernel_size=3, strides=2, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.3)(x)

    x = Conv2D(256, kernel_size=3, strides=2, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.3)(x)

    # Flatten image features to a vector
    flat_image = Flatten()(x)

    # Label input branch (denomination)
    label_input = Input(shape=(1,), dtype='int32', name="denomination_label")
    label_embedding = Embedding(input_dim=num_denominations, output_dim=embed_dim)(label_input)
    label_embedding = Flatten()(label_embedding)

    # Combine image features with label embedding
    combined_features = Concatenate()([flat_image, label_embedding])

    # Fully connected layers for final discrimination
    x = Dense(512)(combined_features)
    x = LeakyReLU(alpha=0.2)(x)
    x = Dropout(0.3)(x)

    # Output validity: probability of being real
    validity_output = Dense(1, activation='sigmoid')(x)

    # Define model with both image and label inputs
    model = Model(inputs=[image_input, label_input], outputs=validity_output)
    return model
```

## Optimizer and Loss Function

```python
BCE_Loss = BinaryCrossentropy(from_logits=False) # Changed 'from_logist' to 'from_logits'

def Genertor_loss(fake_output):
  return BCE_Loss(tf.ones_like(fake_output), fake_output)

def discriminator_loss(real_output, fake_output):
  real_loss = BCE_Loss(tf.ones_like(real_output), real_output)
  fake_loss = BCE_Loss(tf.zeros_like(fake_output), fake_output)  # Corrected to zeros_like for fake loss
  return real_loss + fake_loss
```

```python
generator_opt = Adam(learning_rate=0.0002, beta_1=0.5)
Discrimator_opt = Adam(learning_rate=0.0002, beta_1=0.5)
```

## Training Step Function

```python
EPOCHS = 50  # Number of training epochs
BATCH_SIZE = 64  # Batch size for training
NUM_EXAMPLES_TO_GENERATE = 16 # Number of examples to generate for visualization

# We will reuse this seed overtime (so it's easier)
# to visualize progress in the animated GIF)
seed = tf.random.normal([NUM_EXAMPLES_TO_GENERATE, latent_dim])

# Notice the use of `tf.function`
# This annotation causes the function to be "compiled".
@tf.function
def train_step(images, labels):
    noise = tf.random.normal([BATCH_SIZE, latent_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
      generated_images = denomination_generator([noise, labels], training=True)

      real_output = discriminator([images, labels], training=True)
      fake_output = discriminator([generated_images, labels], training=True)

      gen_loss = Genertor_loss(fake_output)
      disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, denomination_generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_opt.apply_gradients(zip(gradients_of_generator, denomination_generator.trainable_variables))
    Discrimator_opt.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
    return gen_loss, disc_loss
```
